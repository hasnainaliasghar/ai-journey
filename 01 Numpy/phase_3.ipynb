```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("===== Zomato sales analysis =====")
print("\nSlaes data shape: ",sales_data.shape)
print("\n Sample data for 1st 3 restau: ", sales_data[:3])
```

```python
#total sales per year
print("Total sales",np.sum(sales_data[:,1:], axis=0))
print("min sales", np.min(sales_data[:,1:], axis=1))
print("max sales",np.max(sales_data[:,1:], axis=0))
print("Average sales", np.mean(sales_data[:,1:], axis=1))
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

# Average cumulative sales over time

plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Sales Over Time")
plt.xlabel("Time")
plt.ylabel("Sales")
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ", vector1 + vector2)
print("Vector subtraction: ", vector1 - vector2)

print("Multiplication: ", vector1 * vector2)
print("Division: ", vector1 / vector2)

print("Dot product: ", np.dot(vector1, vector2))

cos_theta = np.dot(vector1, vector2) / (
    np.linalg.norm(vector1) * np.linalg.norm(vector2)
)

angle = np.arccos(cos_theta)

print("Angle: ", angle)
```
