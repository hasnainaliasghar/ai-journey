```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
logo = np.load("numpy-logo.npy")
print(logo)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Light/white theme
axes[0].imshow(logo)
axes[0].set_title("Light")
axes[0].axis('off')  # hides grid + ticks

# Dark/inverted theme
dark = 1 - logo
axes[1].imshow(dark)
axes[1].set_title("Dark")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
h, w = logo.shape[:2]

mid = w // 2

dark = 1 - logo
merged = logo.copy()
merged[:, mid:] = dark[:, mid:]
plt.figure(figsize=(10, 5))
plt.imshow(merged)
plt.axis('off')
plt.tight_layout()
plt.show()
```
