```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python


headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/138.0.0.0 Safari/537.36"
    ),
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
}

url = "https://pk.indeed.com/jobs?q=&l=lahore&radius=25"
response = requests.get(url, headers=headers, timeout=15)
print("Status code:", response.status_code)

soup = BeautifulSoup(response.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
jobs = soup.find_all("div",class_ = "job_seen_beacon")
jobs
```

```python
name = []
company = []
city = []
for job in jobs:
    name.append(job.find("span", id= {lambda x : x and x.startswith("jobTitle")}).text.strip())
    company.append(job.find("span", class_="css-19eicqx eu4oa1w0").text.strip())
    city.append(job.find("div", class_ = "css-1f06pz4 eu4oa1w0").text.strip())

```

```python
df = pd.DataFrame(
    {"job_title": name, "company": company, "city": city}
)
```
