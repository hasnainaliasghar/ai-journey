```python
cols = []
for i in df.columns:
    cols.append(i.capitalize())

print(cols)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
df["show_id"].unique()
```

```python
df["show_id"] = df["show_id"].apply(lambda x: x.split("s")[1])
```

```python
df["show_id"]
```

```python
df["show_id"] = pd.to_numeric(df["show_id"], errors="coerce")
```

```python
df.head()
```

```python
df["date_added"] = df["date_added"].apply(lambda x: x.replace("/","-"))
df.head()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

```python
count = df["country"].value_counts().reset_index()
count[count["country"] == "Turkey"]
```

```python
df.to_csv("cleaned.csv", index=False)
```
