## E-commerce Massive DataCleanUp

```python
import pandas as pd
```

```python
pd.__version__
```

```python
df = pd.read_csv("archive/FashionDataset.csv")
```

```python
df.head()
```

```python
"""Deleting Redundant"""
```

```python
df.info()
```

```python
df.shape
```

```python
df.columns
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
"""Renaming the columns"""
for i in df.columns:
    df.rename(columns={ i: i.lower() }, inplace=True)
```

```python
df.columns
```

```python
"""Duplicated rows"""
import matplotlib.pyplot as plt

data = df.duplicated().value_counts()
data.plot(kind='bar')
plt.title("Duplicated Values")
plt.xlabel("Duplicated or Not")
plt.ylabel("Count")
plt.show()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df
```

```python
df.reset_index().drop(columns="index")
```

```python
df.drop(columns="level_0", inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
df
```

```python
df['brandname'] = df["brandname"].apply(lambda x: x.title())
```

```python
df
```

```python
df["mrp"] = df["mrp"].str.replace("\n","",regex=False)
```

```python
df
```

```python
df.head(20)
```

```python
df["mrp"] = df["mrp"].str.replace("Rs","")
df["mrp"] = pd.to_numeric(df["mrp"], errors="coerce")
```

```python
df.head()
```

```python
df["discount"] = df["discount"].str.replace("off","")
```

```python
df.head(50)
```

```python
df.rename(columns={"deatils": "details"},inplace=True)
```

```python
mask = (df["brandname"] == "Nan") & (df["details"] == "Nan")
df = df[~mask]
```

```python
df.reset_index(inplace=True)
```

```python
df.drop(columns=["index"], inplace=True)
```

```python
df.shape
```

```python
df["mrp"] = df["mrp"].fillna(0)
```

```python
df.to_csv("cleaned.csv",index=False)
```
