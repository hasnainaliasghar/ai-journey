```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("anime.csv")
```

```python
df.head()
```

```python
df.loc[1]["Title"]
```

```python
#make a new column for episode count
df["episodes"] = df["Title"].str.extract(r'(\d+\s)')
df["episodes"] = pd.to_numeric(df["episodes"])
df
```

```python
df.loc[1]["episodes"].dtype
```

```python
#make a new column for time stamp
def extract_time(text):
    data = ""
    for i in range(len(text)):
        if text[i] == ')':
            check = True
            for j in range(i+1, i+20):
                data += text[j]

            return data

df["Time"] = df["Title"].apply(extract_time)
df.head()
```

```python
df[['start', 'end']] = df['Time'].str.split(' - ', expand=True)

# Convert to datetime
df['start'] = pd.to_datetime(df['start'], format='%b %Y')
df['end'] = pd.to_datetime(df['end'], format='%b %Y')

# Calculate total months (inclusive)
df['Total_Months'] = (
    (df['end'].dt.year - df['start'].dt.year) * 12 +
    (df['end'].dt.month - df['start'].dt.month) + 1
)
df
```

```python
#which anime has the highest score
df[df['Score'] == df["Score"].max()]
```

```python
#give me top 5 highest scoring anime
df["Title"].head()
```

```python
#which is the longest running anime
df[df["Total_Months"] == df["Total_Months"].max()]
```

```python
#give me top 5 highest scoring anim
df.sort_values(by='Score',ascending=False).head()
```

```python
#which anime has the highest episode count
df[df["episodes"] == df["episodes"].max()]
```

```python
#animes with top 5 episode count
df.sort_values(by = 'episodes', ascending=False).head()
```

```python
df.to_csv("cleaned_data.csv",index=False)
```
